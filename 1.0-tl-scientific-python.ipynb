Plot the color histogram of the rgb image.

Reference: https://matplotlib.org/users/index.html

````python
def plot_color_histogram(image):
    '''
    Plot the histogram of each color channel into a single plot.
    Take a look at the scikit-image submodule skimage.exposure

    image -- RGB image
    '''
    # TODO: implement this function

    try:
        r_hist = skimage.exposure.histogram(image[:, :, 0])
        g_hist = skimage.exposure.histogram(image[:, :, 1])
        b_hist = skimage.exposure.histogram(image[:, :, 2])

        plt.plot(r_hist[1], r_hist[0], color='red', label='Red')
        plt.plot(g_hist[1], g_hist[0], color='green', label='Green')
        plt.plot(b_hist[1], b_hist[0], color='blue', label='Blue')
        plt.xlabel('Intensity')
        plt.ylabel('Frequency')
        plt.title('Color Histogram')
        plt.legend()
        plt.show()
    except Exception as e:
        print(f'Error: {e}')
    
plot_color_histogram(image)
````

## Task 4

Convert the RGB image into the HSV color space. Rotate the colors in the HSV color space, convert it back to the RGB color space and plot it together with its histogram. For this task, you are only allowed to use NumPy.

More information about the HSV color space: https://en.wikipedia.org/wiki/HSL_and_HSV

````python
def rgb2hsv(image):
    '''
    Convert the rgb image into hsv color space and return it.

    image -- rgb image
    '''
    # TODO: implement function
    
    # Ensure float32 format and range [0, 1]
    if image.dtype == np.uint8:
        image = image.astype(np.float32) / 255.0
    elif np.max(image) > 1.0:
        image = image.astype(np.float32) / 255.0
    
    hsv = np.zeros_like(image)
    
    # -- Extract channels
    r, g, b = image[..., 0], image[..., 1], image[..., 2]
    
    # -- Value channel
    v = np.maximum(np.maximum(r, g), b)
    
    # -- Saturation channel
    min_rgb = np.minimum(np.minimum(r, g), b)
    s = np.where(v != 0, (v - min_rgb) / v, 0.)
    
    # -- Hue channel
    delta = v - min_rgb
    
    # Cast to float32 to avoid overflow in numexpr
    delta = delta.astype(np.float32)
    
    # Divide by zero is handled by setting hue to zero
    h = np.zeros_like(delta)
    
    h = np.where(v == r, (g - b) / delta, h)
    h = np.where(v == g, 2. + (b - r) / delta, h)
    h = np.where(v == b, 4. + (r - g) / delta, h)
    h = np.where(delta != 0, h / 6., h)
    
    # Normalize hue to [0, 1]
    h[h < 0] += 1.
    
    # -- Stack channels
    hsv[..., 0] = h
    hsv[..., 1] = s
    hsv[..., 2] = v
    
    return hsv



# test the function
sk_hsv = skimage.color.rgb2hsv(image)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(sk_hsv)
ax[0].set_title('HSV - skimage.color.rgb2hsv')
ax[0].axis('off')
hsv = rgb2hsv(image)
ax[1].imshow(hsv)
ax[1].set_title('HSV - rgb2hsv')
ax[1].axis('off')
plt.show()

def hsv2rgb(image):
    '''
    Convert the hsv image into rgb color space and return it.

    image -- rgb image
    '''
    # TODO: implement function

    # Ensure float32 format and range [0, 1]
    if image.dtype == np.uint8:
        image = image.astype(np.float32) / 255.0
    elif np.max(image) > 1.0:
        image = image.astype(np.float32) / 255.0

    rgb = np.zeros_like(image)

    # -- Extract channels
    h, s, v = image[..., 0], image[..., 1], image[..., 2]

    # -- RGB channels
    i = np.floor(h * 6.) # Integer part
    f = h * 6. - i # Fractional part
    p = v * (1. - s) # Chroma
    q = v * (1. - f * s) # Chroma * (1 - f)
    t = v * (1. - (1. - f) * s) # Chroma * (1 - (1 - f))
    
    # -- Stack channels
    rgb[..., 0] = np.where(i == 0, v, np.where(i == 1, q, np.where(i == 2, p, np.where(i == 3, p, np.where(i == 4, t, v)))))
    rgb[..., 1] = np.where(i == 0, t, np.where(i == 1, v, np.where(i == 2, v, np.where(i == 3, q, np.where(i == 4, p, p)))))
    rgb[..., 2] = np.where(i == 0, p, np.where(i == 1, p, np.where(i == 2, t, np.where(i == 3, v, np.where(i == 4, v, q)))))

    # Normalize to [0, 1]
    rgb = np.clip(rgb, 0., 1.)

    # Cast to uint8
    rgb = (rgb * 255.0).astype(np.uint8)

    return rgb.astype(np.uint8)
    

# test the function 
sk_rgb = skimage.color.hsv2rgb(sk_hsv)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(sk_rgb)
ax[0].set_title('RGB - skimage.color.hsv2rgb')
ax[0].axis('off')
rgb = hsv2rgb(hsv)
ax[1].imshow(rgb)
ax[1].set_title('RGB - hsv2rgb')
ax[1].axis('off')
plt.show()
````
